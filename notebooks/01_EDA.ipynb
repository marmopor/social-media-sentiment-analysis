El dataset contiene mÃºltiples etiquetas emocionales especÃ­ficas (muchas clases).
Debido al tamaÃ±o reducido del dataset (732 muestras) y al fuerte desbalanceo entre clases, se decidiÃ³ agrupar las emociones en tres categorÃ­as generales (Positive, Negative y Neutral) para garantizar estabilidad del modelo y resultados interpretables.

```python
# Eliminamos columnas innecesarias
for col in ["Unnamed: 0", "Unnamed: 0.1"]:
    if col in df.columns:
        df = df.drop(columns=[col])

# Limpieza de la etiqueta
df["Sentiment"] = df["Sentiment"].astype(str).str.strip()

# Mapeo a 3 clases:
positive = {
    "Positive","Joy","Happy","Happiness","Excitement","Contentment",
    "Gratitude","Hope","Hopeful","Love","Pride","Elation","Enthusiasm",
    "Calmness","Serenity","Tranquility","Satisfaction","Relief","Optimism"
}

negative = {
    "Negative","Sad","Sadness","Anger","Fear","Anxiety","Disgust","Hate",
    "Despair","Grief","Loneliness","Frustration","Regret","Shame","Envy",
    "Jealousy","Resentment","Overwhelmed","Exhaustion","Disappointment"
}

def to_3_classes(x):
    if x in positive:
        return "Positive"
    elif x in negative:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment_3"] = df["Sentiment"].apply(to_3_classes)
df["Sentiment_3"].value_counts()
```

```python
# Visualizamos:
plt.figure()
sns.countplot(x="Sentiment_3", data=df)
plt.title("DistribuciÃ³n de clases (3 clases)")
plt.show()
```
